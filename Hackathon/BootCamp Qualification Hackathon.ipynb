## Feature Engineering For the Status Column

```python
def status(x):
  if x =="no":
    return 0
  else:
    return 1
```

```python
#Applying the above function directly to the target column
train.default_status  = train.default_status.apply(status)
```

```python
#Defining X and Y variable
X = train.drop(["default_status","Applicant_ID",'form_field47'], axis =1)
y  = train.default_status
```

```python
# Install catboost using --->  !pip install catboost
```

```python
#Split into train and test set
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
#Creating model 
m2=CatBoostClassifier()

#Training Catboost Model on train set
m2.fit(X_train,y_train)

#Predictiing on Test Set
preds=m2.predict_proba(X_test)
preds = preds[:,1:]

#Evaluating model using roc_auc_score
print("roc: ",roc_auc_score(y_test,preds))
```

```python
#Creating Prediction on test set
p2 = m2.predict_proba(test.drop(["Applicant_ID",'form_field47'], axis =1))
p2 = p2[:,1:]
```

```python
#Initialiazing submission using test set so the size matches
sub = test.copy()
```

```python
#Set target column to prediction 
sub["default_status"] = p2
```

```python
#Create submisssion file so it matches with the default format
test_prediction = sub[["Applicant_ID", 'default_status']]
```

```python
#Pass the createdd file to a csv file
test_prediction.to_csv("first_sub.csv",index= False)
```
